```python
import torchvision.models as models

# Загрузка предобученной ResNet18
model = models.resnet18(pretrained=True)

# Вывод структуры модели
print(model)
```

```python
import torch.nn as nn

num_classes = 10
model.fc = nn.Linear(model.fc.in_features, num_classes)
```

```python
print(model)
```

```python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Преобразования для подготовки изображений
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Размер изображения для ResNet
    transforms.ToTensor(),  # Конвертация в тензор
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),  # Нормализация
])

# Загрузка данных
train_dataset = datasets.ImageFolder('dataset/train', transform=transform)
val_dataset = datasets.ImageFolder('dataset/val', transform=transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
import torch

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
num = 12

# Цикл обучения
for epoch in range(num):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        # Обнуление градиентов
        optimizer.zero_grad()

        # Прямой проход
        outputs = model(inputs)
        loss = criterion(outputs, labels)

        # Обратный проход и шаг оптимизации
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
    print(f'Epoch {epoch+1}/{num}, Loss: {running_loss/len(train_loader)}')

# Валидация
model.eval()
correct = 0,
total = 0
with torch.no_grad():
    for inputs, labels in val_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
print(f'Validation Accuracy: {100 * correct / total:.2f}%')
```

```python
import torch
import torchvision.models as models
import torch.nn as nn

# Инициализация модели ResNet18
model = models.resnet18()
num_classes = 10  # Укажите количество классов (например, 2 для классификации животных)
model.fc = nn.Linear(model.fc.in_features, num_classes)

# Загрузка сохраненных весов
model.load_state_dict(torch.load('resnet18_biology.pth'))
model.eval()  # Переключение модели в режим оценки
```

```python
from PIL import Image
import torchvision.transforms as transforms

# Преобразования для тестового изображения
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Загрузка изображения
image_path = 'Cow_female_black_white.jpg'  # Укажите путь к изображению
image = Image.open(image_path).convert('RGB')  # Открытие изображения
input_tensor = transform(image).unsqueeze(0) 
```

```python
with torch.no_grad():
    output = model(input_tensor)
    _, predicted_class = torch.max(output, 1)

# Отображение результата
class_names = ["Собака", "Лошадь", "Слон", "Бабочка", "Курица", "Кошка", "Корова", "Овца", "Паук", "Белка"]  # Укажите имена классов
print(f'Predicted class: {class_names[predicted_class.item()]}')
```
